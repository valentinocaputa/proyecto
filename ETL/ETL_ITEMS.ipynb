### ETL USER_ITEMS BASANDOME EN LAS FUNCIONES DADAS

```python
#importo librerias

import pandas as pd
import numpy as np
import json
import glob
import ast
import gzip
```

```python
#Procedo a crear el mismo paso que en el archivo user_reviews.json para poder leerlo (no hizo falta descomprimirlo)
info = []     
for i in gzip.open('users_items.json.gz'):
     info.append(ast.literal_eval(i.decode('utf-8')))                                   
     
df_items = pd.DataFrame(info) 
  
df_items_explode= df_items.explode('items')  #convierto columna anidada en un dataframe
df_items_normalizado = pd.json_normalize(df_items_explode['items'])
df_items_explode = df_items_explode.reset_index(drop=True) 
df_items_normalizado = df_items_normalizado.reset_index(drop=True)
df_items = pd.concat([df_items_normalizado, df_items_explode], axis=1) #concateno laa columna normalizada y explorada
df_items = df_items.drop(columns = ['items']) #elimino la columna desanidada anteiormente del dataframe
```
