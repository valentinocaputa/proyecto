### TRANSFORMACIONES COLUMNAS QUE CONTIENEN STRING Y FLOAT

```python

def to_float(value):
    if value is None:
        return np.nan     #creo una funcion que verifica si el valor es none, intente convertir un valor a float, y si falla devuelve nan
    try:
        return float(value)
    except ValueError:
        return np.nan
                         
       
```

```python
#transformo la columna price a float 

df_games['price'].apply(to_float)
```

```python
df_games['price'] = df_games['price'].apply(to_float)
```

```python
df_games['metascore'] = pd.to_numeric(df_games['metascore'], errors='coerce')
```

```python
#realizo la misma funcion para las columnas discount price e id para pasar a float

df_games['id'] = df_games['id'].apply(to_float)

df_games['discount_price'] = df_games['discount_price'].apply(to_float)
```
